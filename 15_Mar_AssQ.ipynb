**Q1. Explain the following with an example:**

    1. Artificial Intelligence
    2. Machine Learning
    3. Deep Learning

Ans -

**1. Artificial Intelligence**
- Al refers to the development of computer systems that can perform tasks that typically require human intelligence.
- It is a smart application that can perform its own task without any human intervention.
- Example:
    - Smart assistants like Siri or Alexa.
    - Recommendation systems like those on Netflix (Movies) or Amazon (Products).
    - Self driving autonomous vehicles car and Robots.


**2. Machine Learning**
- ML is a subset of Al.
- It focuses on developing algorithms that enable computers to learn patterns from data and make decisions without explicit programming.
- It provides Statstool to learn, analyze, visulalize and develop predictive models from the data.
- It also helps in forecasting (Technique used for predicting the future based on the results of previous data).
- Example:
    - Recommendation systems like those on Netflix (Movies) or Amazon (Products).
    - Spam filters in emails that learn to identify and filter out spam based on user feedback without being explicitly programmed for each new spam message.    
    
**3. Deep Learning**
- DL is a subset of ML
- It is basically used to mimic the human brain.
- It involves neural networks with multiple layers (deep neural networks).
- It aims to automatically learn hierarchical representations of data.
- Example:
    - Image and speech recognition systems, where deep neural networks can automatically learn to recognize patterns and features in images or spoken words.
    - Chatbots like Chatgpt.
    - Recommendation systems.

Q2: What is supervised learning? List some examples of supervised learning.

Ans-

**Supervised Learning**

- Supervised Machine Learning is like a student learning with a teacher.
- The computer is taught using labeled examples, where it knows the
correct answers.
- The goal is for the computer to learn and make predictions or decisions
when given new, unseen data.
- In Supervised ML, we know the output of the data set.
    - Classification problems (Categorical outputs)
    - Regression problems (Continuous outputs)
- Predicting if an email is spam or not is a common example of
supervised machine leaming. The algorithm learns from labeled emails
(spam or not) to make predictions on new, unseen emails.
- Teaching a computer to recognize cats and dogs in pictures with
labeled images as a guide is another example of Supervised ML

Q3: What is unsupervised learning? List some examples of unsupervised learning.

Ans-

**Unsupervised learning**

- Unsupervised Machine Learning is like a detective organizing clues
without knowing the answers.
- It creates groups (clusters) of similar things, like customers with
common traits.
- Instead of aiming for a specific outcome, it helps find patterns naturally,
revealing segments in the data.
- In Unsupervised ML, we don't know the output of the data set.
- Here, we don't know any specific output. Instead, we try to group them
into segments.
- Customer segmentation is an example of unsupervised machine
learning, it is the process of dividing a customer base into distinct
groups of individuals that have similar characteristics. This process
makes it easier to target specific groups of customers with tailored
products, services, and marketing strategies

Q5: What are the main differences between supervised, unsupervised, and semi-supervised learning?

Ans -

**Supervised Learning:**

- We teach the model with labeled examples. (Known Output)
- Example: Teaching a computer with labeled images of cats and dogs.

**Unsupervised Learning:**

- No labeled examples; the model discovers patterns on its own. (Unknown Output)
- Example: Sorting a collection of photos into groups without telling the computer what's in each group.

**Semi-Supervised Learning:**

- Uses labeled and unlabeled data. (Mix of Both: Known Output & Unknown Output)
- Example: Teaching a model with some labeled photos of cats and dogs but letting it figure out the rest without labels.

Q6: What is train, test and validation split? Explain the importance of each term.

Ans-

- **Training Data:** Used to teach the model, like giving lessons to a student.

- **Testing Data:** Used to check how well the model learned, like an exam for the model to see if it learned well.

- **Validation Data:** Used to adjust the model for better performance, refining based on feedback.
