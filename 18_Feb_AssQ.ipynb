Q1. What is an API? Give an example, where an API is used in real life.

Ans -

- API stands for Application Programming Interface.
- It is a set of rules that allows different software applications to communicate with each other.
- Example: A weather app uses a weather data API to get the latest weather information and display it. 


Q2. Give advantages and disadvantages of using API.

Ans -
- **Advantages of APIs :** Integration, efficiency, ecosystem growth, scalability, innovation, customization.
- **Disadvantages of APIs :** Security risks, dependency, maintenance overhead, cost, complexity, data privacy concerns.

Q4. Explain REST and SOAP Architecture. Mention shortcomings of SOAP.

Ans -

**REST (Representational State Transfer) :**
- An architectural style for designing networked applications based on a stateless client-server model.
- Uses standard HTTP methods (GET, POST, PUT, DELETE) for communication.
- Lightweight, simple, and easy to implement.
- Widely used in web services and APIs.

**SOAP (Simple Object Access Protocol) :**
- A protocol for exchanging structured information in the implementation of web services.
- Uses XML as the message format and can run over various lower-level protocols.
- Complex and difficult to pick up.
- More rigid and heavyweight compared to REST.

**Shortcomings of SOAP :**
- **Wordiness**: SOAP messages are typically larger due to XML formatting.
- **Slower**: The parsing and processing of XML in SOAP can make it slower.
- **Less Human-Readable**: SOAP messages are less human-readable compared to REST.
- **Complex**: SOAP is complex and difficult to pick up, making it less straightforward to work with.

Q5. Differentiate between REST and SOAP.

Ans -

**REST(Representational State Transfer) :**
- Uses standard HTTP methods.
- Lightweight, simple, and easy to implement.
- Data can be in various formats like JSON, XML.

**SOAP (Simple Object Access Protocol) :**
- Uses XML-based messaging protocol.
- Complex and difficult to pick up.
