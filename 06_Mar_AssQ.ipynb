Q1. What is Statistics?

Ans -

- Statistics is the science of collecting, organizing and analyzing data.
- In data science, statistics is the toolbox of math tricks that we use to understand and make sense of data. 
- It helps us describe what data looks like, figure out patterns, and make predictions or decisions based on that information

Q2. Define the different types of statistics and give an example of when each type might be used.

Ans - There are two types of statistics used in data science.

**1. Descriptive Statistics :-**
- It consists of organizing and summarizing the data.
- It help us to summarize and describe the main aspects of a dataset, like its average, spread, and how its values are distributed.
- It would help you to calculate the average(mean) score and understand the highest and lowest scores in the group.

- Example: Suppose we want to understand the average salary of employees in a company. Using descriptive statistics, we calculate the mean salary to give a sense of the typical earning within the organization.

**2. Inferential Statistics :-**
- It consists of using the data you have measured to form conclusion.
- It used to make predictions or inferences about a larger population based on a sample. 
- For instance, if we survey 500 people to know their opinion about a product, it will help us to draw conclusions about the entire population from this smaller sample.
- Example: Imagine we want to determine if a new drug is effective. You conduct experiments on a sample group and then use inferential statistics to generalize whether the drug will likely have the same effect on a larger population.

Q3. What are the different types of data and how do they differ from each other? Provide an example of
each type of data.

Ans -

Data can be broadly categorized into qualitative and quantitative data.

**Qualitative Data:**
Qualitative data describes qualities or characteristics and is non-numeric, it is further divided into two types.

1. **Nominal Data:** Represents categories without any inherent order. 

    Examples include:
    Colors (e.g., red, blue, green), 
    Types of animals (e.g., dog, cat, bird).

2. **Ordinal Data:** Exhibits a natural order among categories, but the differences between them may not be uniform.

    Examples include:
    Rankings (e.g., 1st place, 2nd place, 3rd place), 
    Levels of education (e.g., elementary, high school, college).
    
**Quantitative Data:**
Quantitative data involves numerical values and is further divided into two types:

1. **Interval Data:** Possesses a consistent and measurable difference between values but lacks a true zero point.

    Examples include:
    Temperature in Celsius or Fahrenheit (e.g., 25°C, 30°C), 
    Years (e.g., 1990, 2005, 2022)
    
2. **Ratio Data:** Shares the characteristics of interval data but includes a true zero point.

    Examples include:
    Height (e.g., 150 cm, 175 cm), 
    Weight (e.g., 50 kg, 75 kg)
    
**The key difference between qualitative and quantitative data lies in their nature: qualitative data describes qualities or attributes, while quantitative data represents quantities or numerical values. Within these categories, the differences lie in the level of measurement (nominal, ordinal, interval, and ratio) and the presence or absence of a true zero point.**

Q5. Explain the concept of levels of measurement and give an example of a variable for each level.

Ans -

- Levels of measurement refer to the different scales or levels at which data can be measured or categorized. 

- Using different types of levels helps in choosing appropriate statistical analyses, understanding data characteristics, and making accurate interpretations, ensuring the right statistical techniques are applied to the data based on its nature and measurement scale.

- There are four main levels of measurement: nominal, ordinal, interval, and ratio.

**Nominal Level:**

- Nominal data represents categories or labels without any inherent order or ranking. It's the simplest form of measurement.
- Example: Eye color (blue, brown, green)

**Ordinal Level:**

- Ordinal data has a natural order or ranking between categories, but the differences between the categories may not be uniform or precisely measurable.
- Example: Educational attainment (high school diploma, bachelor's degree, master's degree, PhD)

**Interval Level:**

- Interval data has a consistent and measurable difference between values, but it lacks a true zero point. The zero point is arbitrary and does not imply the absence of the attribute being measured.
- Example: Temperature in Celsius or Fahrenheit (e.g., 25°C, 30°C)

**Ratio Level:**

- Ratio data has all the characteristics of interval data but includes a true zero point, where zero indicates the absence of the attribute being measured.
- Example: Height (e.g., 150 cm, 175 cm), weight (e.g., 50 kg, 75 kg)

Q9. Describe the difference between descriptive and inferential statistics. Give an example of each
type of statistics and explain how they are used.

Ans -

**Descriptive Statistics :**
- Descriptive statistics summarize and describe the main features of a dataset.
- It consists of organizing and summarizing data.
- e.g. “What is the average height of the students in the entire classroom ?”

**Example :** Imagine you have a basket of different colored candies — red, blue, and green. Descriptive statistics describe what's in the basket. They tell you how many candies of each color there are, the percentage each color makes up, and can show this information visually using a bar chart. It's like counting and showing how many red, blue, and green candies are there.

**Inferential Statistics :**
- Inferential statistics use sample data to make inferences or predictions about a larger population.
- It consists of using the data you have measured to form conclusion.
- e.g. “Are the height of the students in the entire classroom similar to what you except in the entire college ?”

**Example :** Now, imagine you want to know if the whole candy store's basket will have the same proportion of colors as yours. So, you take a small handful from the store's basket. Inferential statistics help decide if your small sample represents the whole store's candies. They estimate the range where the store's proportions might fall based on your sample and try to predict how many candies of each color the store might have. It's like using a small group of candies to guess what the whole store might have in terms of red, blue, and green candies.

Q10. What are some common measures of central tendency and variability used in statistics? Explain
how each measure can be used to describe a dataset.

Ans -

**Measures of central tendency :**
- **Mean :** Describes the average value and it's sensitive to the extreme values.
- **Median :** Describes the middle value, when the dataset is arranged in ascending or descending order and is less affected by extreme values
- **Mode :** Describes the most common value, helpful for understanding the most frequent occurrence.


**Measures of variability :**
- **Range :** The difference between the maximum and minimum values in the dataset. It gives a simple indication of how spread out the data is. 
Provides a quick view of the spread but doesn't consider all values in between
- **Variance :** Measures how much each value in the dataset differs from the mean.
- **Standard Deviation :** The square root of the variance. It represents the average deviation of data points from the mean.
